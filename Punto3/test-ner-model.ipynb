# LOADING NER MODELS

Enlaces a los modelos:

Bert: https://huggingface.co/GetialJuan/bert-base-uncased-finetuned-ner-biobert

Beto: https://huggingface.co/GetialJuan/bert-base-spanish-wwm-cased-finetuned-ner-biobert

Roberta: https://huggingface.co/GetialJuan/xml-roberta-large-finetuned-ner-finetuned-ner-biobert

```python
from transformers import pipeline
```

```python
# bert-base-uncased
bert_base_uncased_pipe = pipeline("token-classification", model="GetialJuan/bert-base-uncased-finetuned-ner-biobert")
```

```python
# BETO
beto_pipe = pipeline("token-classification", model="GetialJuan/bert-base-spanish-wwm-cased-finetuned-ner-biobert")
```

```python
# XML-Roberta
xml_roberta_pipe = pipeline("token-classification", model="GetialJuan/xml-roberta-large-finetuned-ner-finetuned-ner-biobert")
```

# TEST MODELS

```python
text_test = ['Padre fallecio por IAM con 85 años, Madre fallecida con 84 años.',
             'Padre fallecido a los 78 años por ECV isquémica. Madre con HTA y DM2, fallecida a los 82 años.']
```

# TEST BERT-BASE-UNCASED MODEL

```python
# Predicciones
for text in text_test:
  bert_base_uncased_results = bert_base_uncased_pipe(text)

  # Mostrar resultados
  print(f'Texto: {text}')
  for entidad in bert_base_uncased_results:
      print(f"Palabra: {entidad['word']}, Etiqueta: {entidad['entity']}, Puntaje: {entidad['score']:.2f}")

  print("\n")
```

# TEST BETO MODEL

```python
# Predicciones
for text in text_test:
  beto_pipe_results = beto_pipe(text)

# Mostrar resultados
  print(f'Texto: {text}')
  for entidad in beto_pipe_results:
    print(f"Palabra: {entidad['word']}, Etiqueta: {entidad['entity']}, Puntaje: {entidad['score']:.2f}")

  print("\n")
```

# TEST XML-Roberta MODEL

```python

# Predicciones
for text in text_test:
  xml_roberta_results = xml_roberta_pipe(text)

# Mostrar resultados
  print(f'Texto: {text}')
  for entidad in xml_roberta_results:
    print(f"Palabra: {entidad['word']}, Etiqueta: {entidad['entity']}, Puntaje: {entidad['score']:.2f}")

  print("\n")
```
